```python
import numpy as np
import pandas as pd
import jieba
import re

# 导入数据
df = pd.read_csv('adjusted_RandomForestClassifier.csv')
```

```python
df
```

```python
zh_pattern = re.compile(u'[\u4e00-\u9fa5]+')
stop_word = [line.strip() for line in open ('chineseStopWords.txt','r').readlines()]
def extract_words(word):
    word = str(word).strip()
    zh_word = re.findall(zh_pattern,word)
    lst = jieba.cut(''.join(zh_word))
    final_word = [char for char in lst if char not in stop_word]
    return ' '.join(final_word) #空格隔开
data_e = df['message'].apply(lambda x:extract_words(x))   #对整列进行操作
data_extract = pd.DataFrame(data_e.values,columns=['content'])  #转换成列表
```

```python
words = []
for i in data_extract['content']:
    words.append(i)
words
```

### 需要连接情感词典（已修改）

```python
f = open('情感词典/正面评价词语（中文）.txt')
dic_pos = pd.read_table(f,header=None,encoding='UTF-8',delim_whitespace=True)
f = open('情感词典/负面评价词语（中文）.txt')
dic_neg = pd.read_table(f,header=None,encoding='UTF-8',delim_whitespace=True)
f = open('情感词典/正面情感词语（中文）.txt')
dic_pos = pd.concat([dic_pos,pd.read_table(f,header=None,encoding='UTF-8',delim_whitespace=True)])
f = open('情感词典/负面情感词语（中文）.txt')
dic_neg = pd.concat([dic_neg,pd.read_table(f,header=None,encoding='UTF-8',delim_whitespace=True)])
dic_pos.columns = ['words']
dic_neg.columns = ['words']
dic_pos
```

```python
words_filter = []
for w in words:
    ws = w.split(' ')
    dic_w = []
    # 遍历每一个词
    for i in range(len(ws)):
        now_w = ws[i]
        if now_w in dic_pos.values or now_w in dic_neg.values:
            dic_w.append(now_w)
    words_filter.append(dic_w)
words_filter
```

```python
# 共现矩阵
node_dict = {} # 节点字典，节点名+频数
edge_dict = {} # 边字典，起点+目标点+频数
# 遍历每一句评论
for ws in words_filter:
    # 遍历每一个词
    for i in range(len(ws)):
        now_w = ws[i]
        if now_w not in node_dict:
            node_dict[now_w] = 1
        else:
            node_dict[now_w] += 1
        # 取出其他词
        ws_s = ws[i:]
        # 统计当前词与其之后的词出现情况
        for w_s in ws_s:
            A,B = now_w,w_s
            if A == B:
                continue
            # 转化词序
            if A > B:
                A,B = B,A
            # 生成边
            key = A+','+B
            if key not in edge_dict:
                edge_dict[key] = 1
            else:
                edge_dict[key] += 1
```

```python
# 转化成DataFrame
df = pd.DataFrame([node_dict]).T.reset_index()
df.columns = ['Label','Weight']
df.sort_values(by='Weight',ascending=False,inplace=True)
df = df.iloc[:200]
df.to_csv('node.csv',index=None)
```

```python
# 转化成DataFrame
df1 = pd.DataFrame([edge_dict]).T.reset_index()
df1.columns = ['Label','Weight']
```

```python
#拆分
df1['Source'] = df1['Label'].map(lambda x:x.split(',')[0])
df1['Target'] = df1['Label'].map(lambda x:x.split(',')[1])
df1
```
