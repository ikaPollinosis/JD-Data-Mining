```python
import numpy as np
import pandas as pd
import jieba
import re

# 导入数据
df = pd.read_csv('adjusted_RandomForestClassifier.csv')
```

```python
df
```

```python
zh_pattern = re.compile(u'[\u4e00-\u9fa5]+')
stop_word = [line.strip() for line in open ('chineseStopWords.txt','r').readlines()]
def extract_words(word):
    word = str(word).strip()
    zh_word = re.findall(zh_pattern,word)
    lst = jieba.cut(''.join(zh_word))
    final_word = [char for char in lst if char not in stop_word]
    return ' '.join(final_word) #空格隔开
data_e = df['message'].apply(lambda x:extract_words(x))   #对整列进行操作
data_extract = pd.DataFrame(data_e.values,columns=['content'])  #转换成列表
```

```python
cts = []
for i in data_extract['content']:
    cts.append(i)
cts
```

```python
# TFIDF
from sklearn.feature_extraction.text import TfidfVectorizer

# 使用模型
vectorizer = TfidfVectorizer(max_features=700,max_df=0.5,smooth_idf=True)
X = vectorizer.fit_transform(cts)
X.toarray()
```

```python
vectorizer.vocabulary_
```

```python
# 奇异值分解
from sklearn.decomposition import TruncatedSVD

#实例化、训练
svd_model = TruncatedSVD(n_components=5,algorithm='randomized',n_iter=100,random_state=32451)
X_topic = svd_model.fit_transform(X)
X_topic
```

```python
terms = vectorizer.get_feature_names()
```

```python
for i in range(5):
    # 取出主题词权重
    comp = svd_model.components_[i]
    # 转化字典
    dz = dict(zip(terms,comp))
    dz_s = sorted(dict(zip(terms,comp)).items(),key=lambda x:x[1],reverse=True)[:10]
    print('Topic' +str(i)+':'+' '.join([d[0]for d in dz_s]))
```

### 总结

```python
np.argmax(X_topic,axis=1)
```

```python
# 计算每个文本属于哪个主题
np.argmax(X_topic,axis=1)
keywords = []
for i in np.argmax(X_topic,axis=1):
    comp = svd_model.components_[int(i)]
    # 转化字典
    dz = dict(zip(terms,comp))
    dz_s = sorted(dict(zip(terms,comp)).items(),key=lambda x:x[1],reverse=True)[:10]
    keywords.append(' '.join([d[0]for d in dz_s]))
```

```python
keywords
```

```python
df['LSA_Topic'] = np.argmax(X_topic,axis=1)
df['LSA_Keywords'] = keywords
df.to_csv('LSAResult.csv',index=0,columns=None,encoding='utf_8')
df
```

```python
ds = pd.DataFrame(xs)
ds['label']=np.argmax(X_topic,axis=1)
ds
```

```python
import matplotlib.pyplot as plt
labels = list(set(list(ds['label'])))
colors = ['skyblue','tomato','pink','purple','red','green','brown','yellow','gold','orange']
for i in range(len(labels)):
    ds_ = ds[ds['label']==labels[i]]
    plt.plot(ds_[0],ds_[1],'o',color = colors[i])
plt.show()
```
